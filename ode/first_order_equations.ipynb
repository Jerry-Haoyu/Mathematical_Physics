# First-order Equations


We first consider the general form of first-order equations $y'=f(y,t),y(0)=y_0$, where $y(0)=y_0$ is a necessary constraint that select the correct trajectory out of all possible trajectories. Such problem is known as initival value problem. Two important aspects:

:::{note} Slogan
(1) It's more than sufficient to guarantee the existence of a solution if $f\in C^1$, futhermore we get that $y$ depends continuously on the inital condition $y_0$ 

(2) There is no general closed-form *analytical* formula for first-order equations

(3) There is however, general *numerical/analytical* algorithms
:::


:::{prf:theorem} Existence and Uniqueness
Consider an inital value problem:

\begin{align} y' &=f(y,t)\\
y(0) &= y_0
\end{align}

if $f$ is *locally Lipchitz continuous* in $y$ and *continuous* in a box $R=[-a,a]\times [-b,b]$ $t$ then there exists a unique solution $y$ to the IVP in $t\in[-a,a]$. 
:::

:::{prf:proof}
We use a construction *Picard's Approximation* to compelete the proof. First we reduce the problem into an equivalent integral form. 

**Claim**: Suppose exists $\phi(t)$ with $\phi(0)=\phi_0$ that statisfies the integral equation below, than $y=\phi(t)$ satsifies the IVP and vice versa
$$
\phi(t)-\phi_0=\int_0^t f(\phi(s),s)ds
$$
To show the forward direction we just use Fundamental theorem of calculus. The reverse direction is just direct integration from $t=0$ to arbitrary $t$. 
Picard algorithm allow us to compute an infinite sequence of functions which approximates the solution $y=\phi(t)$ better and better. First we define the metric space we 
work in, define the set:
$$
\mathcal Y = \{\phi: [-a,a]\to [-b,b] : \phi(0)=0\}
$$
